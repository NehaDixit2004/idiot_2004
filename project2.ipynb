```python
import matplotlib.pyplot as plt

# categorical columns detect karo
cat_cols = df.select_dtypes(include='object').columns

# har column ka bar graph
for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f"{col} distribution")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
df['person_gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# categorical columns select
cat_cols = df.select_dtypes(include='object').columns

# sabka count combine karo
combined_counts = pd.Series(dtype=int)

for col in cat_cols:
    counts = df[col].value_counts()
    counts.index = col + " - " + counts.index.astype(str)   # label add
    combined_counts = pd.concat([combined_counts, counts])

# ek hi bar graph
plt.figure(figsize=(12,6))
combined_counts.plot(kind='bar')
plt.title("Complete Dataset Categorical Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# categorical columns select
cat_cols = df.select_dtypes(include='object').columns

# sabka count combine karo
combined_counts = pd.Series(dtype=int)

for col in cat_cols:
    counts = df[col].value_counts()
    counts.index = col + " - " + counts.index.astype(str)
    combined_counts = pd.concat([combined_counts, counts])

# pie chart
plt.figure(figsize=(8,8))
combined_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Complete Dataset Categorical Distribution (Pie Chart)")
plt.ylabel("")   # extra label remove
plt.show()
```
