Que2. Write a program that takes a list of elements and creates a dictionary showing the frequency of each element

```python
lst = [1, 2, 2, 3, 1, 4, 2, 3, 3, 3]

frequency = {}
for el in lst:
    if el in frequency:
       continue
    else:
        frequency[el] = lst.count(el)
print(frequency)
```

Que4 :Write a program to merge two dictionaries and resolve common keys by keeping the larger value.


```python

dict1 = {"a": 10, "b": 20, "c": 30}
dict2 = {"b": 25, "c": 15, "d": 40}
merged = {}

for key in set(dict1) | set(dict2):
    merged[key] = max(dict1.get(key, 0), dict2.get(key, 0))

print(merged)
```

Que5 :Using a dictionary comprehension, create a dictionary from a list of integers where keys are numbers and values are their cubes, but only for odd numbers.

```python
numbers = [21, 22,23,24,25,26,27,28,29]

cube_dict = {n: n**3 for n in numbers if n % 2 != 0}

print(cube_dict)
```
